### Setup

```python
import pandas as pd
import numpy as np

school_path = "Resources/schools_complete.csv"
student_path = "Resources/students_complete.csv"

school_data = pd.read_csv(school_path)
student_data = pd.read_csv(student_path)

school_data_complete = pd.merge(student_data, school_data, how="left", on=[
                                "school_name", "school_name"])

#school_data.head()
student_data.head()
#school_data_complete.head()
```

# District Summary

#### Total number of schools

```python
school_count = len(school_data_complete["school_name"].unique())
# school_count
```

#### Total number of students

```python
student_count = len(school_data_complete["Student ID"])
# student_count
```

#### Total Budget


```python
# Used "school_data" df to get correct budget 
total_budget = school_data["budget"].sum()
# total_budget
```

#### Average Math Score

```python
avg_math_score = school_data_complete["math_score"].mean()
# avg_math_score
```

#### Average Reading Score

```python
avg_reading_score = school_data_complete["reading_score"].mean()
# avg_reading_score
```

#### Overall Passing Rate

```python
overall_passing_rate = (((avg_math_score + avg_reading_score) / 2) /100)
# overall_passing_rate
```

#### % of Students Passing Math

```python
# finding amount of students that had 70 or greater in math
passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70, :]
total_passing_math = len(passing_math)

# passing math / total amount of students
percentage_passing_math = total_passing_math / student_count
percentage_passing_math = percentage_passing_math
# percentage_passing_math
```

#### % of Students Passing Reading

```python
# finding amount of students that had 70 or greater in reading
passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70, :]
total_passing_reading = len(passing_reading)

# passing reading / total amount of students
percentage_passing_reading = total_passing_reading / student_count
percentage_passing_reading = percentage_passing_reading
# percentage_passing_reading
```

## Complete: District Summary   

```python
district_summary = pd.DataFrame({
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": avg_math_score,
    "Average Reading Score": avg_reading_score,
    "% Passing Math": percentage_passing_math,
    "% Passing Reading": percentage_passing_reading,
    "% Overall Passing Rate": overall_passing_rate,
})

# district_summary
```

### Reformatting District Summary

```python
district_summary = district_summary.style.format({
    "Total Students": "${:,}",
    "Total Budget": "${:,}",
    'Average Math Score': "{:.1f}",
    'Average Reading Score': "{:.1f}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "% Overall Passing Rate": "{:.1%}"})

district_summary
```

# School Summary

```python
by_school_df = school_data_complete.groupby(["school_name"])

# by_school_df.head() 
```

#### School Types

```python
type_per_school = school_data.set_index("school_name")["type"]
# type_per_school
```

#### Total Students per School

```python
student_per_school = school_data.set_index("school_name")["size"]
# student_per_school
```

#### Total Budget per School

```python
budget_per_school = school_data.set_index("school_name")["budget"]
# budget_per_school
```

#### Budget per Student

```python
per_student_budget = budget_per_school / student_per_school
# per_student_budget
```

#### Average Math Score per School

```python
school_avg_math_score = by_school_df["math_score"].mean()
# school_avg_math_score
```

#### Average Reading Score per School

```python
school_avg_reading_score = by_school_df["reading_score"].mean()
# school_avg_reading_score
```

####  Passing Math  % per School

```python
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID']
school_percentage_passing_math = pass_math.count()/ student_per_school

# school_percentage_passing_math
```

#### Passing Reading  % per School

```python
pass_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID']
school_percentage_passing_reading = pass_reading.count()/ student_per_school

# school_percentage_passing_reading
```

#### Overall Passing Rate per School

```python
school_percentage_passing_overall = (school_percentage_passing_math + school_percentage_passing_reading) / 2

# school_percentage_passing_overall
```

## Complete: School Summary

```python
school_summary = pd.DataFrame({
    "School Type": type_per_school,
    "Total Students": student_per_school,
    "Total School Budget": budget_per_school,
    "Per Student Budget": per_student_budget,
    "Average Math Score": school_avg_math_score,
    "Average Reading Score": school_avg_reading_score,
    "% Passing Math": school_percentage_passing_math,
    "% Passing Reading": school_percentage_passing_reading,
    "% Overall Passing Rate": school_percentage_passing_overall,
})

# school_summary.head()
```

### Reformatting School Summary

```python
school_summary.style.format({
    "Total Students": '{:,}',
    "Total School Budget": "${:,}",
    "Per Student Budget": "${:.0f}",
    "Average Math Score": "{:.1f}",
    "Average Reading Score": "{:.1f}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "% Overall Passing Rate": "{:.1%}"
})
```

# Top Performing Schools (By Passing Rate)

```python
top_five_schools = school_summary.sort_values(["% Overall Passing Rate"], ascending=False)

top_five_schools.head().style.format({
    "Total Students": '{:,}',
    "Total School Budget": "${:,}",
    "Per Student Budget": "${:.0f}",
    "Average Math Score": "{:.1f}",
    "Average Reading Score": "{:.1f}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "% Overall Passing Rate": "{:.1%}"
})
```

# Bottom Performing Schools (By Passing Rate)

```python
bottom_five_schools = school_summary.sort_values(["% Overall Passing Rate"])

bottom_five_schools.head().style.format({
    "Total Students": '{:,}',
    "Total School Budget": "${:,}",
    "Per Student Budget": "${:.0f}",
    "Average Math Score": "{:.1f}",
    "Average Reading Score": "{:.1f}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "% Overall Passing Rate": "{:.1%}"
})
```

# Math Scores by Grade

```python
ninth_math = student_data.loc[student_data['grade'] == "9th"].groupby("school_name")[
    "math_score"].mean()
tenth_math = student_data.loc[student_data['grade'] == "10th"].groupby(
    "school_name")["math_score"].mean()
eleventh_math = student_data.loc[student_data['grade'] == "11th"].groupby(
    "school_name")["math_score"].mean()
twelfth_math = student_data.loc[student_data['grade'] == "12th"].groupby(
    "school_name")["math_score"].mean()

math_scores = pd.DataFrame({
    "9th": ninth_math,
    "10th": tenth_math,
    "11th": eleventh_math,
    "12th": twelfth_math
})

math_scores
```

# Reading Score by Grade

```python
ninth_reading = student_data.loc[student_data['grade'] == "9th"].groupby("school_name")[
    "reading_score"].mean()
tenth_reading = student_data.loc[student_data['grade'] == "10th"].groupby(
    "school_name")["reading_score"].mean()
eleventh_reading = student_data.loc[student_data['grade'] == "11th"].groupby(
    "school_name")["reading_score"].mean()
twelfth_reading = student_data.loc[student_data['grade'] == "12th"].groupby(
    "school_name")["reading_score"].mean()

reading_scores = pd.DataFrame({
    "9th": ninth_reading,
    "10th": tenth_reading,
    "11th": eleventh_reading,
    "12th": twelfth_reading
})

reading_scores
```

# Scores by School Spending

```python
school_spending_bins = [0, 585, 615, 645, 675]
school_spending = ["< $585", "$585 - 615", "$615 - 645", "$645 - 675"]

school_summary["Per Student Budget"] = pd.cut(
    per_student_budget, school_spending_bins, labels=school_spending)

scores_spending_math = school_summary.groupby(
    ["Per Student Budget"]).mean()["Average Math Score"]
scores_spending_reading = school_summary.groupby(
    ["Per Student Budget"]).mean()["Average Reading Score"]
scores_spending_math_percentage = school_summary.groupby(
    ["Per Student Budget"]).mean()["% Passing Math"]
scores_spending_reading_percentage = school_summary.groupby(
    ["Per Student Budget"]).mean()["% Passing Reading"]
scores_spending_overall_percentage = (
    scores_spending_math + scores_spending_reading) / 2


scores_by_school_spending_summary = pd.DataFrame({
    "Average Math Score": scores_spending_math,
    "Average Reading Score": scores_spending_reading,
    "% Passing Math": scores_spending_math_percentage,
    "% Passing Reading": scores_spending_reading_percentage,
    "% Overall Passing Rate": scores_spending_overall_percentage,
})

scores_by_school_spending_summary
```

# Scores by School Size

```python
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_summary["School Size"] = pd.cut(school_summary["Total Students"], size_bins, labels=group_names)

scores_size_math = school_summary.groupby(
    ["School Size"]).mean()["Average Math Score"]
scores_size_reading = school_summary.groupby(
    ["School Size"]).mean()["Average Reading Score"]
scores_size_math_percentage = school_summary.groupby(
    ["School Size"]).mean()["% Passing Math"]
scores_size_reading_percentage = school_summary.groupby(
    ["School Size"]).mean()["% Passing Reading"]
scores_size_overall_percentage = (
    scores_size_math + scores_size_reading) / 2

scores_by_school_size_summary = pd.DataFrame({
    "Average Math Score": scores_size_math,
    "Average Reading Score": scores_size_reading,
    "% Passing Math": scores_size_math_percentage,
    "% Passing Reading": scores_size_reading_percentage,
    "% Overall Passing Rate": scores_size_overall_percentage
})

scores_by_school_size_summary






```
